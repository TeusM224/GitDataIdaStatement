```python
import pandas as pd
import numpy as np
import requests as rq 
import json
import pymysql
```

```python
url="https://finances.worldbank.org/resource/ebmi-69yj.json"
response1 = rq.get(url).json()
```

```python
response=rq.get(url)
dictresponse = json.loads(response.text)
dictresponse
```

```python
df=pd.DataFrame(response1)
df
```

```python
#Elimino la columna "End of period"
df.drop(['end_of_period'], axis=1, inplace=True)
df
```

```python
df.isna().sum()
df.country.value_counts()
df.info()
```

```python
df['borrower']=df['borrower'].apply(lambda x: x.upper())
```

```python
 #Eliminamos service_charge_rate porque en todas nos dan el mismo valor pero en un nuevo df por si necesitamos el original
df1=df.drop(['service_charge_rate'], axis=1)
```

```python
df1.drop(['agreement_signing_date'], axis=1, inplace=True)
df1.drop(['board_approval_date'], axis=1, inplace=True)
```

```python
#Reemplazo valores nan por NONE
df1.replace(np.nan, None, inplace=True)
```

FORMAS DE SUBIR EL DATAFRAME A MYSQL

```python
#Creo una lista de las columnas
campos = ['credit_number', 'region', 'country_code', 'country', 'borrower',
       'credit_status', 'currency_of_commitment', 'project_id', 'project_name',
       'original_principal_amount', 'cancelled_amount', 'undisbursed_amount',
       'disbursed_amount', 'repaid_to_ida', 'due_to_ida',
       'exchange_adjustment', 'borrower_s_obligation', 'sold_3rd_party',
       'repaid_3rd_party', 'due_3rd_party', 'credits_held',
       'first_repayment_date', 'last_repayment_date',
       'effective_date_most_recent_', 'closed_date_most_recent_',
       'last_disbursement_date']
```

```python
#Creo una lista que guardará las tuplas con los valores respectivos de cada campo(columna)
lista = []
for fila in range(0,len(df1)):
    lista.append(tuple(df1[campos].loc[fila]))
```

```python
#Realizo la conexión con el server y la DATABASE
conexion=pymysql.connect(
    host='localhost',
    database='ida_statement',
    user='root',
    password='Lohare01*'
)
cursor = conexion.cursor()
```

```python
#Realizo la query que inserta en la base de datos las filas y columnas
query='INSERT INTO ida_statement_credits_grants (credit_number, region, country_code, country, borrower, credit_status, currency_of_commitment, project_id, project_name, original_principal_amount, cancelled_amount, undisbursed_amount, disbursed_amount, repaid_to_ida, due_to_ida, exchange_adjustment, borrower_s_obligation, sold_3rd_party, repaid_3rd_party, due_3rd_party, credits_held, first_repayment_date, last_repayment_date, effective_date_most_recent_, closed_date_most_recent_, last_disbursement_date) VALUES(%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s)'
cursor.executemany(query,lista)
conexion.commit()
conexion.close()
```
